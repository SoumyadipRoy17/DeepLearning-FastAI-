```python

!pip install -q transformers
from transformers import pipeline

# One unified model and pipeline
nlp = pipeline("text2text-generation", model="google/flan-t5-small")

# 1. Feature Extraction (simulated)
print("Feature Extraction:", nlp("List important keywords from: 'Transformers are revolutionizing NLP.'")[0]['generated_text'])

# 2. Fill-Mask (simulated)
print("Fill-Mask:", nlp("Fill in the blank: The capital of Germany is <mask>.")[0]['generated_text'])

# 3. Question Answering
qa_prompt = "Context: The Nile is the longest river in the world. Question: What is the longest river in the world?"
print("Question Answering:", nlp(qa_prompt)[0]['generated_text'])

# 4. Sentence Similarity
similarity_prompt = ("Are these two sentences similar?\n"
                     "Sentence 1: The car is parked in the garage.\n"
                     "Sentence 2: The automobile is inside the garage.")
print("Sentence Similarity:", nlp(similarity_prompt)[0]['generated_text'])

# 5. Summarization
print("Summarization:", nlp("Summarize: Artificial Intelligence enables machines to mimic human intelligence.")[0]['generated_text'])

# 6. Table Question Answering (simulated via text)
table_prompt = ("Given the table: Year | Revenue\n2020 | $5M\n2021 | $7M\n"
                "Question: What was the revenue in 2021?")
print("Table QA:", nlp(table_prompt)[0]['generated_text'])

# 7. Text Classification
print("Text Classification:", nlp("Classify the sentiment of: 'I hated the movie. It was too long and boring.'")[0]['generated_text'])

# 8. Text Generation
print("Text Generation:", nlp("Write a short story about a lonely robot learning emotions.")[0]['generated_text'])

# 9. Text Ranking (simulated)
ranking_prompt = ("Rank these statements by relevance to climate change:\n"
                  "1. Global warming affects sea levels.\n"
                  "2. Dogs are friendly animals.\n"
                  "3. Carbon emissions cause global temperature rise.")
print("Text Ranking:", nlp(ranking_prompt)[0]['generated_text'])

# 10. Token Classification (NER simulated)
print("Token Classification:", nlp("Extract named entities from: 'Google was founded by Larry Page and Sergey Brin in California.'")[0]['generated_text'])

# 11. Translation
print("Translation:", nlp("Translate to French: 'The future of AI is exciting.'")[0]['generated_text'])

# 12. Zero-Shot Classification
print("Zero-Shot Classification:", nlp("Determine if the following is positive, negative, or neutral: 'I guess it's okay.'")[0]['generated_text'])
```
